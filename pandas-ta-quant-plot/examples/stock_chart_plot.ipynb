```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
from pandas_ta_quant import pd
from pandas_ta_quant_plot.plot_utils import color_positive_negative
from pandas_ta_quant_plot import Hist
```

```python
df = pd.read_csv("../pandas_ta_quant_plot_test/.data/SPY.csv", index_col="Date", parse_dates=True)
df.tail()
```

```python
with df.ta.plot(range_slider_price='Close', backend='nbAgg', start=-120) as p:
    p["main"].candlestick("Open", "High", "Low", "Close")
    p["main"].line(p.df["Close"])
    p["volume"].bar("Volume", colors=color_positive_negative(p.df))
```

```python
with df.ta.plot(range_slider_price='Close', backend='nbAgg') as p:
    p["main"].candlestick("Open", "High", "Low", "Close")
    p["main"].line(p.df["Close"]) > Hist(bins=100)
    p["volume"].bar("Volume", colors=color_positive_negative(p.df)) > Hist(bins=20)
```
